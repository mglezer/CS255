```python
def to_hex_str(n):
    return '{:02x}'.format(n)

def str_to_list(string):
    return [ord(c) for c in string]

def hex_str_to_list(hex_str):
    return [int(hex_str[i:i+2], 16) for i in range(0, len(hex_str), 2)]

def to_ascii(list):
    return [chr(x) if x >= 32 and x <= 122 else '?' for x in list]

def is_alpha(n):
    return (n >= 65 and n <= 90) or (n >= 97 and n <= 122)

def xor(l1, l2):
    return [l1[i] ^ l2[i] for i in range(min(len(l1), len(l2)))]

to_ascii(to_char_list("656667"))
```

```python
# Examine the xor of every pair of ctexts.
xors = [[] for i in range(len(ctexts))]
print(xors)
for i in range(len(ctexts)):
    for j in range(len(ctexts)):
        xors[i].append(xor(ctexts[i], ctexts[j]))
```

```python
# If a character in an xor is in a-z or A-Z, one input must have been a space, and the other the alphabetical character with reverse capitalization (e.g. A -> a).
# To determine which is which, we can take each cipher text in that position and see which produces more alphabetical characters when xor'd with other ciphertexts.
# Whichever produces more is the space. Once we know which is the space, we can xor the ciphertext with space to get the key value in that position.

counts = [0] * 100
key = [0] * 100
for i in range(len(xors)):
    for k in range(min_len):
        count = 0
        for j in range(len(xors)):    
            if is_alpha(xors[i][j][k]):
                count += 1
        if count > counts[k]:
            counts[k] = count
            key[k] = ctexts[i][k] ^ 32
                
```
