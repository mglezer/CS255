```python
N = mpz("""
6484558428080716696628242653467722787263437207069762630604390703787 \
9730861808111646271401527606141756919558732184025452065542490671989 \
2428844841839353281972988531310511738648965962582821502504990264452 \
1008852816733037111422964210278402893076574586452336833570778346897 \
15838646088239640236866252211790085787877""".replace(" \n", ""))

rt = mpz(gmpy2.ceil(gmpy2.sqrt(N)))

# Starting at the root, make a guess about A.
# If we guess A correctly, then we know x = sqrt(A^2 - N), and (A+x)*(A-x) = N.

found = False
for A in range(rt, rt + mpz(1) << 20):
    x = mpz(gmpy2.sqrt(A**2 - N))
    if (A-x)*(A+x) == N:
        print(x)
        print((A-x)*(A+x) == N)
        print(A-x, A+x)
        found = True
        break
if not found:
    print("Could not find a solution!")
```

```python
# This time, we know that the numerator of A is ODD, which means A must end in .5.
# We have also proven that A is within 1/(2 + sqrt(6)) = 0.23 of sqrt(6*N). These
# two facts together mean that A and sqrt(6*N) will always have the same whole number part,
# but A will end in 0.5.
A = gmpy2.floor(gmpy2.sqrt(6*N)) + 0.5
```

```python
print(p, q)
```
